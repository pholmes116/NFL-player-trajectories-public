```python
import pandas as pd
```

# Games

Game data: The games.csv contains the teams playing in each game. 

The key variable is gameId.

```python
games_df = pd.read_csv(f"{folder}games.csv")
games_df.head()
```

# Player Play

Player play data: The player_play.csv file contains player-level stats for each game and play. 

The key variables are gameId, playId, and nflId.

```python
player_play_df = pd.read_csv(f"{folder}player_play.csv")
print(player_play_df.columns.values)
player_play_df.head()
```

# Players

Player data: The players.csv file contains player-level information from players that participated in any of the tracking data files. 

The key variable is nflId.

```python
players_df = pd.read_csv(f"{folder}players.csv")
players_df.head()
```

```python
players_df.value_counts(
    subset=['position']
    ).reset_index(
        name='count'
        ).sort_values(
            by='count', ascending=False)
```

```python
players_df[players_df['position'] == 'LB'].head()
```

```python
players_df[players_df['position'] == 'DB'].head()
```

# Play

Play data: The plays.csv file contains play-level information from each game. 

The key variables are gameId and playId.

```python
plays_df = pd.read_csv(f"{folder}plays.csv")
plays_df.head()
```

```python
plays_df.columns.values
```

```python
plays_df[plays_df['playNullifiedByPenalty']=="N"].shape[0]/plays_df.shape[0]
```

```python
plays_df.value_counts(
    subset=['penaltyYards']
    ).reset_index(
        name='count'
        ).sort_values(
            by='count', ascending=False)
```

```python
plays_df.value_counts(
    subset=['gameClock']
    ).reset_index(
        name='count'
        ).sort_values(
            by='count', ascending=False)
```
